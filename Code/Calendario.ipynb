#### Es finde? 1 si, 0 no

```python
weekend = ['Friday','Saturday', 'Sunday']
list_day = []
for i in list_calendar:
    if calendar.day_name[datetime.strptime(i, '%d-%m-%Y').weekday()] in weekend:
        list_day.append(1)
    else: 
        list_day.append(0)
list_day
```
